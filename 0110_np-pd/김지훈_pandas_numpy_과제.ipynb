#### NaN, 데이터 타입, 개수 확인

```python
data.info()
```

#### birthDate 컬럼으로 age 컬럼 추가하기

##### 1) birthDate 컬럼의 자료형을 object에서 datetime으로 전환

```python
from dateutil.parser import parse

data['birthDate'] = data['birthDate'].map(parse, na_action = 'ignore')
data.head(30)
```

##### 2) age 컬럼 생성, apply 및 lambda 함수를 써서 나이 계산

```python
from datetime import datetime as dt

td = dt.now()
data['age'] = data['birthDate'].apply(lambda x: td.year - x.year - ((td.month, td.day) < (x.month, x.day)))
data['age'].head(30)
```

##### 3) age 결측치를 선수 평균 나이로 대체

```python
data['age'] = data['age'].fillna(data['age'].agg('mean'))
data['age'].tail(30)
```

#### height 컬럼(feet-inch 형식)으로 heightCm 컬럼 추가하기

##### 1) heightCm 컬럼 생성, apply 및 lambda 함수를 사용해서 height 컬럼의 feet과 inch를 cm로 변환 ('-'를 기준으로 split해서 각각을 cm로 변환해 더하기)



```python
data['heightCm'] = data['height'].apply(lambda l: float(l.split('-')[0]) * 30.48 + float(l.split('-')[1]) * 2.54)
data['heightCm'].head(30)
```
